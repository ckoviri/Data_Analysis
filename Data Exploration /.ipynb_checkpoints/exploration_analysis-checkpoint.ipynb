# Data Wrangling
## Data Gathering

```python
# downloading the zip files 2018 jan, feb, march data from source link
dataset_year = ["201801","201802","201803"]
if not os.path.exists("source_file"):
        os.makedirs("source_file")
for year in dataset_year:
    response = requests.get(f"https://s3.amazonaws.com/fordgobike-data/{year}-fordgobike-tripdata.csv.zip")
    with open(f"source_file/{year}.csv.zip", mode = "wb") as file:
        file.write(response.content)
```

```python
# UnZip the data
files = list(os.listdir("./source_file"))
for file in files:
    if ".zip" in file:
        with zfile.ZipFile(f"./source_file/{file}",'r') as zipfiles:
            zipfiles.extractall("./")
```

## Cleaning

### Convert dtypes of the columns
- 'start_time'to datetime
- 'end_time' to datetime
- 'member_birth_year' to int

```python
# dtype conversion code
df["start_time"] = pd.to_datetime(df.start_time)
df["end_time"] = pd.to_datetime(df.end_time)
df["member_birth_year"]= df.member_birth_year.round().astype('Int64')
```

```python
#test
df[["start_time","end_time","member_birth_year"]].dtypes
```

#### Treating missing values 
- Fill the missing values with mean of the column

```python
#plot to find the null values
plt.figure(figsize =(8,8))
plt.title("Missing value plot")
plt.xlabel("columns")
plt.ylabel("row numbers")
sns.heatmap(df.isnull());
```

#### Filling null values 

As the member_birth_year is continues data we can replace the null values with mean and the null values of member_gender with "No data" 

```python
#getting mean value year
df["member_birth_year"].mean().round()
```

```python
# fillling null values
df["member_birth_year"].fillna(df["member_birth_year"].mean().round(),inplace = True)
df["member_gender"].fillna("No data",inplace = True)
```

```python
#test
df.isnull().sum()
```

## Extracting New Columns
- Day, month column using Start time and end time
- Creating new column by converting duration seconds to mintues
- Age of member using member birth year
- Distance using longitude and latitude

```python
#extracting day column
df["start_day"]=df['start_time'].dt.day_name()
df["start_month"]=df['start_time'].dt.month_name()
df["end_day"]=df['end_time'].dt.day_name()
df["end_month"]=df['end_time'].dt.month_name()
```

```python
#test
df[["start_day","start_month","end_day","end_month"]]
```

#### Extracting minitues from duration seconds

```python
# converting seconds to minitues 
df['duration_min'] = (df['duration_sec']/60).astype(int)
```

```python
#test
df['duration_min']
```

#### Extracting age from duration member year

```python
# calculating Age from member birth year
df['member_age'] = 2020 - df['member_birth_year']
```

```python
#test
df['member_age']
```

```python
df['member_age'].describe()
```

```python
df.info()
```

```python
# code from https://stackoverflow.com/questions/40452759/pandas-latitude-longitude-to-distance-between-successive-rows
lat1 = df.start_station_latitude
lon1= df.start_station_longitude
lat2=   df.end_station_latitude
lon2 = df.end_station_longitude
earth_radius=6371
lat1, lon1, lat2, lon2 = np.radians([lat1, lon1, lat2, lon2])
a = np.sin((lat2-lat1)/2.0)**2 + \
np.cos(lat1) * np.cos(lat2) * np.sin((lon2-lon1)/2.0)**2
df['dist'] =  earth_radius * 2 * np.arcsin(np.sqrt(a))
```

## Univariate Exploration

> In this section, we are going to investigate distributions of individual variables, unusual points or outliers and prepare the data to look at relationships between variables.

```python
plt.hist(df["duration_min"],bins =100)
plt.title("Ford Go Bike system rides duration in minutes")
plt.xlabel('Duration in minutes');
plt.ylabel('No. of Bike Rides');
plt.show();
sns.boxplot(df["duration_min"]);
plt.title("Box plot for Ford Go Bike system rides duration in minutes");
```

Duration in mins data is right skewed beacuse of spread of data till 1400. We can consider the extreme data as outliers but there might be a chance of users who took trip for 1400miles or user may took the bike but didnt ride. So, we cannot consider this data as outliers.

### Binning Duration column 
from above plot and observation. its better to binning the data for better understanding. So i binned the data as 4 groups. Trips below 'below-5min', between '10-30mins',between '30-60mins' and'above60mins.

```python
df["duration_min"].describe()
```

```python
bin_labels= ['below-10min', '10-30mins', '30-60mins', 'above60mins']
bins = [0,10,30,60,1439]
df["duration"] = pd.cut(df['duration_min'], bins,labels=bin_labels)
```

```python
chart = sns.catplot(data = df, x = 'duration', kind = 'count', aspect = 1.7);
chart.set_axis_labels('Duration', 'Number of Bike Trips');
chart.fig.suptitle("Duration vs Number of Bike Trips ");
```

Fromt the above plot. It's clear that most of the rides are below 30mins in which most of them are below 10mins.
